```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore' )

dataset = pd.read_csv("Merged data set.csv")
```

```python
dataset.drop('Date/Time', axis=1, inplace=True)
```

```python
#Method 1
knn_dataset1 = (dataset - np.min(dataset))/(np.max(dataset) - np.min(dataset))
knn_dataset1
```

```python
#Method 2
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler().fit(dataset)
knn_dataset2 = mms.transform(dataset)
knn_dataset2 = pd.DataFrame(knn_dataset2)
knn_dataset2
```

```python
x = knn_dataset1.drop("Active power", axis = 1)
y = knn_dataset1["Active power"] 
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 2)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor().fit(x_train, y_train)
y_pred = knn_model.predict(x_test)
```

```python
predict_df = pd.DataFrame({"y_Test" : y_test, "y_Predicted" : y_pred})
predict_df.head()
```

```python
predict_df = (predict_df*(np.max(dataset.displacement) - np.min(dataset.displacement))) + np.min(dataset.displacement)
predict_df.head()
```
