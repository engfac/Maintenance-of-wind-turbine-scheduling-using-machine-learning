```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore') #to filter warning messages
```

```python
df=pd.read_csv('H_data.csv')
df.set_index("Time", inplace=True)
df.head()
```

```python
# Split data to features and target
from sklearn.model_selection import train_test_split
x=df.drop(['Active power'],axis=1)

# Feature scaling
from sklearn.preprocessing import StandardScaler
std_x=StandardScaler()
X=std_x.fit_transform(x)


y=df['Active power'].values.ravel()
#y=y.reshape(-1,1)


# Test Train split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=2)
```

```python
from sklearn.neural_network import MLPRegressor
Model_MLP = MLPRegressor(hidden_layer_sizes=(100,100,100),activation="relu" ,solver="adam",learning_rate="constant",alpha=0.0001, max_iter=100)
Model_fit = Model_MLP.fit(x_train, y_train)
```

```python
y_pred_MLP = Model_MLP.predict(x_test)
data_1 = pd.DataFrame({'Predicted value': y_pred_MLP, 'Real Value': y_test})
data_1.head(20)
```

```python
from sklearn.metrics import accuracy_score,r2_score,mean_absolute_error, mean_squared_error

print('R2-score:',r2_score(y_test,y_pred_MLP))
print('RMSE:',np.sqrt(mean_squared_error(y_test,y_pred_MLP)))
print('MAE:',mean_absolute_error(y_test, y_pred_MLP))
print('MSE:',mean_squared_error(y_test, y_pred_MLP))
```
